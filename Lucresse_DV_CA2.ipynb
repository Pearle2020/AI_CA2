### Code Implementation

```python
#this code is an updated  version of the code provided to us in class , some changes were made in order to adapt to our requierements

import heapq 
import time
def calculate_distances(graph_fig1, starting_vertex, end_vertex): 
    
    
    
    distances = {vertex: float('infinity') for vertex in graph_fig1}
    distances[starting_vertex] = 0
   
    previous = {}


    pq = [(0, starting_vertex)]
    while len(pq) > 0:
        current_distance, current_vertex = heapq.heappop(pq)
        if current_vertex == end_vertex:
            break

        # Nodes can get added to the priority queue multiple times. We only
        # process a vertex the first time we remove it from the priority queue.
        if current_distance > distances[current_vertex]:
            continue

        for neighbor, weight in graph_fig1[current_vertex].items():
            distance = current_distance + weight

            # Only consider this new path if it's better than any path we've
            # already found.
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                previous[neighbor] = current_vertex
                heapq.heappush(pq, (distance, neighbor))
    
                
    path = []
    current_node = end_vertex
    while current_node:
        path.append(current_node)
        current_node = previous.get(current_node, None)

    path.reverse()
    
    return path ,distances[end_vertex]
   
   
```

```python
import time
# to identify the running time of the Dijsktra Algorithm
start = time.time()
print(f"the shortest path from A to S is: {calculate_distances(graph_fig1, 'A','S')}")
end =time.time()
runningTime=  end - start
print(f"The running time of the Dijsktra's Algorithm is: {runningTime}")
```

#### A* Algorithm
   The A* search algorithm is an algorithm that helps in finding the shortest path between a start vertex and a end vertex on a weighted graph  by taking in consideration the cost of the visited vertices from the start vertex. This algorithm uses Heuristic to estimates a numerical values that will be added to the cost of the current  vertices.
  the formula used in this algorithm is:
####  f(n) = g(n) + h(n)
Where:
- f(n) : is the Calculated total cost of path

- g(n): is The cost of path between the first node and the current node

- h(n): is Heuristic 
 the determination of the heuristic
   
#### A* Implementation

```python
from collections import deque
import time

class Graph:

    def __init__(self, adjacency_list):
        self.adjacency_list = adjacency_list

    def get_neighbors(self, v):
        return self.adjacency_list[v]
    #the heuristic function, base on the the direct line from A to the each vertices
    # in our case the heuristic what determine using a ruler in our whiteboard
    def h(self, n):
        He = {
    'A': 0 ,
    'B': 1.7,
    'C': 1,
    'D': 3,
    'E': 4.1,
    'F': 3,
    'G': 1.9,
    'H': 1.7,
    'K': 1.6,
    'J': 3.5,
    'L': 3.3,
    'N': 5,
    'M': 5,
    'P': 6.4,
    'Q': 5.4,
    'W': 5,
    'V': 6.5,
    'S': 7,
    'U': 7.9,
    'T': 8.4,
    'R': 7.5
     }
        return He[n]
    def a_star(self, start, end):
        open_list = set([start])
        closed_list = set([])
        g = {}# this has the current distant from the starting node(this are initialize to infinity)

        g[start] = 0

        # parents contains an adjacency map of all nodes
        parents = {}
        parents[start] = start

        while len(open_list) > 0:
            n = None

            # find a node with the lowest value 
            for v in open_list:
                if n == None or g[v] + self.h(v) < g[n] + self.h(n):
                    n = v;

            if n == None:
                print('Path does not exist!')
                return None
               

            # if the current node is the stop_node
            # then we begin reconstructin the path from it to the start_node
            if n == end:
                reconst_path = []

                while parents[n] != n:
                    reconst_path.append(n)
                    n = parents[n]

                reconst_path.append(start)
                #this build the shortest path by backtracting
                reconst_path.reverse()

                print('The shortest Path from A to S is : {}'.format(reconst_path))
                return reconst_path
            

            # for all neighbors of the current node do
            for (m, weight) in self.get_neighbors(n):
                
                if m not in open_list and m not in closed_list:
                    open_list.add(m)
                    parents[m] = n
                    g[m] = g[n] + weight

                else:
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        parents[m] = n

                        if m in closed_list:
                            closed_list.remove(m)
                            open_list.add(m)

            # remove n from the open_list, and add it to closed_list
            # because all of his neighbors were inspected
            open_list.remove(n)
            closed_list.add(n)
      

        print('Path does not exist!')
        
        return None
```

#### Understanding the state of the algorithm
  For a little understnding of the algorithm,
we start with
- the Initialization state:
in this state of the algorithm, the open and start list are being initialize. the open node being our starting vertex and the the closed list initialize empty. Here the distance is also being initialise to 0 wich is the distance between the starting vertex and itself.

- the looping state
here we are looking for the vertex with the lowest f(n) value stored in our open list.

- the reconstruction state
here we check if the current vertex is our end vertex, if yes we provide the path to it , if not we go back to our looping state.

- neighbors check
in this state , we evaluate the adjacente node to the current vertex, if they arre in our open list, we update their F(n) value, if it they are in the close list we move back to  the looping state and if they are not in the close or open list we add then to our open list.
